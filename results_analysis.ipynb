```python
def load_data(filepath):
    data = np.loadtxt(filepath)
    return data
```

```python
def plot_single():

    plt.figure(figsize=(10, 6))

    filepath = './Remi-results/Results/H+/v=0.1/Cross/cross.inp'
    data = load_data(filepath)
    
    x = data[:, 0]  # Column 1
    y = data[:, 0] * data[:, 2]  # $1 * $3 in Gnuplot

    cubic_spline = CubicSpline(x, y)
    x_new = np.linspace(x.min(), x.max(), 300)  
    y_new = cubic_spline(x_new) 
    # y_new = np.maximum(y_new, 0)

    from scipy.interpolate import PchipInterpolator

    # Replace the cubic spline section with PCHIP fitting
    pchip_interp = PchipInterpolator(x, y)
    y_new_pchip = pchip_interp(x_new)

    # Clamping negative values to 0
    # y_new_pchip = np.maximum(y_new_pchip, 0)


    if 'Straight-line' not in filepath: 
        # plt.plot(x,y)
        plt.scatter(x, y)
        plt.plot(x_new, y_new)    
        plt.plot(x_new, y_new_pchip)
 

        # plt.rc('axes', prop_cycle=(cycler('color', plt.cm.Set1.colors)))  # Use a predefined colormap
    # else: 
    #     plt.scatter(x, y, linestyle=linestyle, linewidth=3, marker='o') 
    #     plt.plot(x_new, y_new, linestyle=linestyle, linewidth=3)  
    #     # plt.rc('axes', prop_cycle=(cycler('color', plt.cm.Set1.colors)))    
      
    plt.xlim(0,10)
    sns.set_theme(style='whitegrid')  # Set seaborn theme
    plt.xlabel('b (a.u.)')  
    plt.ylabel('bP') 
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_h_fig3(ax):
    # ax.figure(figsize=(10, 6))

    for filepath, label, linestyle, color in files_and_labels_H_fig3:
        data = load_data(filepath)
    
        x = data[:, 0]  # Column 1
        y = data[:, 0] * data[:, 2]  # $1 * $3 in Gnuplot

        mask = x <= 10
        x_filtered = x[mask]
        y_filtered = y[mask]

        cubic_spline = CubicSpline(x_filtered, y_filtered)
        x_new = np.linspace(x_filtered.min(), x_filtered.max(), 500)  
        y_new = cubic_spline(x_new) 
        y_new = np.maximum(y_new, 0)

        if 'Straight-line' not in filepath: 
            ax.scatter(x_filtered, y_filtered, color=color)
            ax.plot(x_new, y_new, linestyle=linestyle, linewidth=3, color=color,label=label)       
        # else: 
        #     ax.scatter(x_filtered, y_filtered, linestyle=linestyle, color=color) 
        #     ax.plot(x_new, y_new, linestyle=linestyle, linewidth=3, color=color)  
      
    # ax.set_xlim(0,10)
    sns.set_theme(style='whitegrid')  
    ax.tick_params(axis='both', which='major', labelsize=16) 
    # ax.set_xlabel('b (a.u.)', fontsize = 16)  
    ax.set_ylabel(r'bP$_{CX}$', fontsize = 18) 
    ax.text(0,4.8, r'H${^+}$+H($1s$)', fontsize = 20)
    ax.text(10,4.8, 'A', fontweight='bold', fontsize=20)
    # ax.text(0.1,2.3, r'He${^2+}$+H($1s$)', fontsize = 20)
    # ax.text(0.1,1.8, r'Li${^3+}$+H($1s$)', fontsize = 20)
    # ax.text(0.1,4.1, r'Be${^4+}$+H($1s$)', fontsize = 20)
    # ax.legend(fontsize = 14)
    ax.xaxis.set_ticks_position('both')  
    ax.yaxis.set_ticks_position('both')
    ax.tick_params(axis='x', direction='in', length=8) 
    ax.tick_params(axis='y', direction='in', length=8)
    ax.grid(False)
    # ax.savefig('fig3_H.pdf', format='pdf', dpi=300)
    # ax.savefig('fig3_H.ps', format='ps')
    # ax.show()
    ax.set_xticklabels([])
    ax.xaxis.set_minor_locator(AutoMinorLocator(2)) 
    ax.yaxis.set_minor_locator(AutoMinorLocator(2))  
    ax.tick_params(axis='x', which='minor', direction='in', length=4, color='gray') 
    ax.tick_params(axis='y', which='minor', direction='in', length=4, color='gray')  
```
